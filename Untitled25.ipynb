# 1.a Logistic Regression

```python
# Import the dataset
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Import the dataset
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Ms data analytics Groups/January-Session/Classification Algorithms/Social_Network_Ads.csv")
# Content of the dataset
dataset.head()

# Select the columns
# Select the specific columns
X = dataset.iloc[:, 2:4]
y = dataset.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
# Preprocessing step
st = StandardScaler()
X =  st.fit_transform(X)

# Importing the model
from sklearn.linear_model import LogisticRegression
lg_reg = LogisticRegression()
# Applying the model
lg_reg.fit(X, y)

# Calculating the score of the model
lg_reg.score(X, y)
y_pred = lg_reg.predict(X)
```

```python
lg_reg.score(X, y)
```

# Visualization of the model

```python
from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.pyplot as plt

# Assuming X and y are properly defined and lg_reg is your trained logistic regression model
# X should be a 2D NumPy array or equivalent containing two features
# y should be a NumPy array containing target labels (e.g., 0 and 1)

# Define the grid for decision boundary visualization
num_points = 100  # Adjust as needed
X1, X2 = np.meshgrid(
    np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, num_points),
    np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, num_points),
)



# Flatten the grid and make predictions for each point
grid_points = np.array([X1.ravel(), X2.ravel()]).T
predictions = lg_reg.predict(grid_points).reshape(X1.shape)

# Plot the decision boundary
plt.contourf(
    X1,
    X2,
    predictions,
    alpha=0.75,
    cmap=ListedColormap(["red", "green"])  # Ensure valid color names
)

# Plot the data points
for i, j in enumerate(np.unique(y)):
    plt.scatter(
        X[y == j, 0],
        X[y == j, 1],
        c=["red", "green"][i],  # Map colors to class labels
        label=f"Class {j}"
    )

# Add labels, title, and legend
plt.title("Logistic Regression")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.legend()
plt.show()
```

# 1. b Descision Tree

```python
# Import the dataset
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Import the dataset
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Ms data analytics Groups/January-Session/Classification Algorithms/Social_Network_Ads.csv")
# Content of the dataset
dataset.head()
```

```python
X = dataset.iloc[:, [2,3]].values
Y = dataset.iloc[:, 4].values
```

# Splitting the data into Training and Testing

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                    Y,
                                                    test_size = 0.25,
                                                    random_state=0)
```

# Feature Scaling

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Building the classifier

```python
#Build our Classifier
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, Y_train)
#Predicting The Results
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_pred, Y_test)
print("The accuracy is: {}%".format(acc*100))
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_pred, Y_test)
print(cr)
```

# Visuliazation of the Model

```python
# Import necessary libraries
from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.pyplot as plt

# Visualising the Training set results
x_set, y_set = X_train, Y_train

# Creating the grid of minimum and maximum values from X_train
X1, X2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1,
                               stop = x_set[:, 0].max() + 1,
                               step = 0.01),
                    np.arange(start = x_set[:, 1].min() - 1,
                              stop = x_set[:, 1].max() + 1,
                              step = 0.01))

# Plotting the decision boundary (classifier line)
plt.contourf(X1,
             X2,
             classifier.predict(np.array([X1.ravel(),
                                          X2.ravel()]).T).reshape(X1.shape),
            alpha = 0.4,
            cmap = ListedColormap(['red', 'green']))  # Corrected color format

# Plotting the datapoints in red and green
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0],
                x_set[y_set == j, 1],
                c = ListedColormap(['red', 'green'])(i),  # Corrected color format
                label = j,
                s = 8,
                marker = '*'
               )

# Set the limits for the axes
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Add titles and labels
plt.title("Desicion Tree Classifier")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import the dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Ms data analytics Groups/January-Session/Kmeans/Mall_Customers.csv")
# Content of the dataset
df.head()
```

```python
# Extracting the useful features from the dataset
X = df.iloc[:, [3, 4]].values
```

# This algorithm use dendogram to find out the optimal number of cluster

# Kmean use elbow method

```python
# Finding the optimal no.of Clusters by Dendrograms
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
# Apply the model of clustering
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=5, linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
# Plotting the results
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster5')
plt.title('Hierarchical Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
