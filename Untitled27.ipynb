Handle Anomalies

 Detect Anomalies (Outliers)

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
df_outliers = df[(z_scores > 3).any(axis=1)]
print(df_outliers)
```

Option 3: Impute (replace) with median or mean

```python
# Check the column names in your DataFrame to verify the column you want to impute
print(df.columns)

# Replace 'Adj Close' with the actual name of the column you want to impute.
# Define the threshold value that makes sense for your data and analysis.
column_to_impute = 'Adj Close' # Example: replace with the actual column name
threshold = df[column_to_impute].quantile(0.95) # Example: Use the 95th percentile as a threshold

# Calculate the median of the specified column
median_val = df[column_to_impute].median()

# Replace values in the specified column that are greater than the threshold with the median value
df.loc[df[column_to_impute] > threshold, column_to_impute] = median_val

# You might want to check the changes
# print(df[column_to_impute].describe())
```
