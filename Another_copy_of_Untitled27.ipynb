CREATING A TIME BASED FEATURES E.G (DAY OF THE WEEK MONTH SEASON)

Step 1: Load the Dataset

```python
import pandas as pd

# Load the dataset
# Ensure the file 'Tesla_Stock_Prices.csv' is in the same directory as this notebook
# If the file is in a different directory, provide the correct path, e.g., 'path/to/your/data/Tesla_Stock_Prices.csv'
try:
    df = pd.read_csv('Tesla_Stock_Prices.csv')

    # Convert the 'Date' column to datetime objects
    # Replace 'Date' with the actual name of your date column if it's different
    df['Date'] = pd.to_datetime(df['Date'])

    # Check the data types after conversion
    print(df.info())

except FileNotFoundError:
    print("Error: 'Tesla_Stock_Prices.csv' not found.")
    print("Please ensure the file is in the correct directory or update the path in the code.")
```

Create Time-Based Features

Basic time components:

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['dayofweek'] = df['Date'].dt.dayofweek   # Monday=0, Sunday=6
df['hour'] = df['Date'].dt.hour
df['minute'] = df['Date'].dt.minute
df['second'] = df['Date'].dt.second
```

 Week and quarter:

```python
df['weekofyear'] = df['Date'].dt.isocalendar().week
df['quarter'] = df['Date'].dt.quarter
```

Weekend

```python
df['is_weekend'] = df['Date'].dt.dayofweek >= 5
```

PART OF THE DAY

```python
def get_part_of_day(hour):
    if hour < 6:
        return 'Night'
    elif hour < 12:
        return 'Morning'
    elif hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'

df['part_of_day'] = df['hour'].apply(get_part_of_day)
```

Time since / until

```python
df['time_since_start'] = df['Date'] - df['Date'].min()
df['days_since_start'] = df['time_since_start'].dt.days
```

Set the Date as Index

```python
df.set_index('Date', inplace=True)
```

Resample or Rolling Features

```python
# Resample to daily data (useful for time series aggregation)
# Select only numerical columns for mean calculation
numerical_df = df.select_dtypes(include=np.number)
daily_df = numerical_df.resample('D').mean()

# Rolling average over 7 days
# Replace 'your_numeric_column' with the actual name of a numerical column you want to calculate the rolling average for
# For example, to calculate the rolling mean of 'Adj Close':
df['rolling_7d'] = df['Adj Close'].rolling(window=7).mean()
```
