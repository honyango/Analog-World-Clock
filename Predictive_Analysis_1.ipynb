```python
import numpy as np
from scipy.stats import pearsonr

# Define the variables x and y with some sample data
x = np.array([2,5,1,3,4,1,5,3,4,2])
y = np.array([50,57,41,54,54,38,63,48,59,49])

# Calculate Pearson correlation
correlation, p_value = pearsonr(x, y)

print(f"Pearson correlation coefficient: {correlation:.3f}")
print(f"P-value: {p_value:.3f}")
```

Linear Regression Analysis in Python

```python
import numpy as np
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression

# Define the variables x and y with some sample data
x = np.array([2,5,1,3,4,1,5,3,4,2])
y = np.array([50,57,41,54,54,38,63,48,59,49])
# Calculate Pearson correlation
correlation, p_value = pearsonr(x, y)
print(f"Pearson correlation coefficient: {correlation:.3f}")
print(f"P-value: {p_value:.3f}")

# Linear Regression Analysis in Python
# Reshape x for sklearn
x_reshaped = np.array(x).reshape(-1, 1)
y_array = np.array(y)

# Create model and fit
model = LinearRegression()
model.fit(x_reshaped, y_array)

# Get parameters
slope = model.coef_[0]
intercept = model.intercept_
print(f"Regression equation: y = {slope:.2f}x + {intercept:.2f}")
```

PLot Regression Line

```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
# Predicted values
y_pred = model.predict(x_reshaped)

# Plot
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('Number of TV Commercials (x)')
plt.ylabel('Store Sales (y)')
plt.title('Linear Regression: TV Commercials vs. Store Sales')
plt.legend()
plt.grid(True)
plt.show()
```

Write and explain a Python code for a regression analysis of the data set in question.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Step 1: Prepare the data
x = np.array([1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]).reshape(-1, 1)
y = np.array([100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100])

# Step 2: Create and train the model
model = LinearRegression()
model.fit(x, y)

# Step 3: Make predictions
y_pred = model.predict(x)

# Step 4: Evaluate the model
r2 = r2_score(y, y_pred)
print(f"R Score: {r2:.4f}")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Slope: {model.coef_[0]:.2f}")

# Step 5: Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression Analysis')
plt.xlabel('TV Commercials (x)')
plt.ylabel('Store Sales (y)')
plt.legend()
plt.grid(True)
plt.show()
```

QUADRATIC REGRESSION FIT

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
# Data
x = np.array([1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]).reshape(-1, 1)
y = np.array([100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100])

# Transform to include polynomial terms
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

# Fit model
model = LinearRegression()
model.fit(x_poly, y)

# Predict
y_pred = model.predict(x_poly)
# Plot
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', label='Quadratic Fit')
plt.xlabel('Number of Commercials')
plt.ylabel('Sales')
plt.title('Quadratic Regression Fit')
plt.legend()
plt.show()
```

3. Choose a multi-dimensional data set (with at least 3 data features/attributes) suitable for a binary classification problem. Use Support Vector Machines (SVM) to build a binary classification model. Evaluate its performance and perform a 3D visualization of the data classes and the boundary between them.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from mpl_toolkits.mplot3d import Axes3D

# Load dataset - Corrected to load the standard iris dataset
iris = datasets.load_iris()

# Use the features and target from the loaded iris dataset
X = iris.data  # Features (e.g., sepal length, sepal width, petal length, petal width)
y = iris.target  # Target labels (0, 1, or 2 for the three iris species)

# For binary classification, we need to select two classes. Let's choose the first two classes (0 and 1).
# Find indices of the first two classes
indices = np.where((y == 0) | (y == 1))[0]
X = X[indices, :]
y = y[indices]

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
#print output
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
# Now you can continue with building and evaluating the SVM model for binary classification.
# For example:
# model = svm.SVC(kernel='linear')
# model.fit(X_train, y_train)
# y_pred = model.predict(X_test)
# print("Accuracy:", accuracy_score(y_test, y_pred))
# print(classification_report(y_test, y_pred))

# Note: 3D visualization will require selecting only 3 features from the dataset.
```

Step 2: Train an SVM Model

```python
# Create SVM classifier with RBF kernel
clf = svm.SVC(kernel='rbf', C=1.0, gamma='scale', probability=True)
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

Step 3: 3D Visualization with Division Boundary

```python
# Create a meshgrid for plotting the decision boundary
h = 0.2  # Reduce step size for a smoother boundary
x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
z_min, z_max = X[:, 2].min()-1, X[:, 2].max()+1

xx, yy, zz = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h),
                         np.arange(z_min, z_max, h))

# Flatten the grid to pass to classifier
grid = np.c_[xx.ravel(), yy.ravel(), zz.ravel()]

# Calculate the decision function for each point in the grid
# We use try-except because decision_function might fail for all points if they are outside the training data range
try:
    decision = clf.decision_function(grid)
    decision = decision.reshape(xx.shape)

    # Extract the points where the decision function is close to 0
    # These points form the decision boundary surface
    boundary_points = grid[np.abs(decision.ravel()) < h*5] # Select points where absolute decision value is small

    # Plot
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Plot original data points
    scatter = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='coolwarm', s=60, edgecolors='k', label='Data Points')

    # Plot decision boundary points
    if len(boundary_points) > 0:
        ax.scatter(boundary_points[:, 0], boundary_points[:, 1], boundary_points[:, 2],
                   color='red', s=10, alpha=0.5, label='Decision Boundary')
    else:
        print("No boundary points found within the meshgrid range.")


    ax.set_xlabel('Sepal Length') # Use correct feature names for Iris dataset
    ax.set_ylabel('Sepal Width')
    ax.set_zlabel('Petal Length')
    ax.set_title('SVM Classification with Decision Boundary (3D)')
    ax.legend()
    plt.show()

except Exception as e:
    print(f"Error during decision boundary calculation or plotting: {e}")
    print("Ensure your meshgrid range covers the data and the trained model is appropriate for 3D.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from mpl_toolkits.mplot3d import Axes3D

# 1. Load dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Use only two classes: Setosa and Versicolor
binary_mask = y < 2
X = X[binary_mask]
y = y[binary_mask]

# Select 3 features for 3D visualization
X = X[:, :3]  # e.g., sepal length, sepal width, petal length

# 2. Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Train SVM
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# 4. Predict and Evaluate
y_pred = svm_model.predict(X_test)
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")

# 5. 3D Visualization
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
# Plot points
scatter = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='coolwarm', s=60, edgecolor='k')

# Axes labels
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
ax.set_title('3D View of Iris Data (Setosa vs Versicolor)')

# Create legend
legend_labels = [target_names[0], target_names[1]]
ax.legend(handles=scatter.legend_elements()[0], labels=legend_labels, loc="best")
plt.show()
```
